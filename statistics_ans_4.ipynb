```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example dataset
data = {
    'CGPA': [5, 6, 7.5, 8, 9, 9.5],
    'PACKAGE': [10, 12, 8, 15, 40, 50]
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
```

```python

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['CGPA'], df['PACKAGE'])
plt.title('Scatter Plot of CGPA Vs PACKAGE')
plt.xlabel('CGPA')
plt.ylabel('PACKAGE')
plt.grid(True)
plt.show()
```

```python
# Calculate correlation coefficient
correlation_coefficient = np.corrcoef(df['CGPA'], df['PACKAGE'])[0, 1]
print("Correlation coefficient:", correlation_coefficient)
```
